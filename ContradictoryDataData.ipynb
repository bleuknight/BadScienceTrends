```python
import pandas as pd

#this is the file which links contains all query terms
terms = pd.read_csv("queryTerms.csv")
queryTerms = list(terms['term1'])
```

```python
queryTerms
```

```python
contradictoryData = ('"contradictory+Data"[TIAB] AND "' + Q + '"[TIAB]' for Q in queryTerms)
contradictoryDataList = ['"contradictory+Data"[TIAB] AND "' + Q +'"[TIAB]' for Q in queryTerms]
print(contradictoryDataList)
```

```python
from Bio import Entrez
Entrez.email = "scientistbleu@gmail.com"
contradictoryDataCount = []
for r in contradictoryData:
    handle = Entrez.esearch(db = "pubmed", term = r, retmax = 100000)
    record = Entrez.read(handle)
    count = record["Count"]
    contradictoryDataCount.append(count)
print(contradictoryDataCount)
```

```python
countContradictoryData = dict(zip(queryTerms, contradictoryDataCount))
countContradictoryData
```

```python
from Bio import Entrez
Entrez.email = "happylittleaccident@gmail.com"
#creates open dictionary to store PMID with query term
contradictoryDataID = {}
def getContradictoryDataIDs(QTs):
    for Q in QTs:
        t = '"contradictory+Data"' +'[TIAB] AND "' + Q + '"[TIAB]'
        handle = Entrez.esearch(db = "pubmed", term = t, retmax = 100000)
        record = Entrez.read(handle)
        PMIDS = record["IdList"]
        contradictoryDataID[Q] = PMIDS

getContradictoryDataIDs(queryTerms)
print(contradictoryDataID)
```

```python
df_contradictoryData = pd.DataFrame([countContradictoryData, contradictoryDataID]).T
df_contradictoryData.columns = ['d{}'.format(i) for i, col in enumerate(df_contradictoryData, 1)]
df_contradictoryData
```

```python
df_contradictoryData.to_csv("contradictoryData.csv")
```

```python
contradictoryData_PMIDs = set().union(*contradictoryDataID.values())
len(contradictoryData_PMIDs)
```

```python
contradictoryData_abstracts = {}

#creates list to store entries without abstract text
without_abstract_contradictoryData = []

Entrez.email = 'scientistbleu@gmail.com'

handle = Entrez.efetch(db="pubmed", id=','.join(map(str, contradictoryData_PMIDs)),
                       rettype="xml", retmode="text")
records = Entrez.read(handle)
for pubmed_article in records['PubmedArticle']:
    pmid = int(str(pubmed_article['MedlineCitation']['PMID']))
    article = pubmed_article['MedlineCitation']['Article']
    if 'Abstract' in article:
        abstract = article['Abstract']['AbstractText'][0]
        contradictoryData_abstracts[pmid] = abstract
    else:
       without_abstract_contradictoryData.append(pmid)

print(contradictoryData_abstracts)
```

```python
import json
with open('contradictoryData.txt', 'w') as outfile:
    json.dump(contradictoryData_abstracts, outfile)
with open('contradictoryData_noAbs.txt', 'w') as outfile:
    json.dump(without_abstract_contradictoryData, outfile)
```
